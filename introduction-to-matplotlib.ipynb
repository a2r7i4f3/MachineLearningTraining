```python
# make plot from dictionary

corona_virus_death = {"Italty": 7000,
                     "Japan": 3000,
                     "Bangladesh": 7,
                     "China": 6000}

fig, ax = plt.subplots()
ax.bar(corona_virus_death.keys(), corona_virus_death.values()) # bar plot
ax.set(title = 'Corona Cases Deaths',
      ylabel = 'Deaths',
      xlabel = 'country')
```

```python
fig, ax = plt.subplots() 
# show bar horizontally 
ax.barh(list(corona_virus_death.keys()), list(corona_virus_death.values()))
```

# Tow option for subplots

```python
# option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                            ncols = 2,
                                            figsize = (10,5))

ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(corona_virus_death.keys(), corona_virus_death.values())
ax4.hist(np.random.randn(1000));
```

```python
# option 2

fig, ax = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize = (10,5))

ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].hist(np.random.randn(10000));
ax[1, 1].barh(list(corona_virus_death.keys()), list(corona_virus_death.values()))
```

## Advantages and differences between pyplot and matplotlib. which one is better?


* when plotting something quickly, pylot is good to use
* when plotting something more advanced, user the oo method . oo = object oriented

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
over_50.target.values
```

```python
over_50.chol.values
```

```python
over_50.ndim, type(over_50), len(over_50)
```

```python
# pyplot method

over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             figsize = (10, 6))
```

```python
# oo method mixed with pyplot method

fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             ax = ax)

#ax.set_xlim([45, 100])
```

```python
# oo method from scratch

fig, ax = plt.subplots(figsize = (10, 6))

#plot the data
scatter = ax.scatter(x = over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target']);

#Customize
ax.set(title = 'Heart Disease and Cholesterol Levels',
       xlabel = 'Age',
       ylabel = 'Cholesterol')

# Add a legend
ax.legend(*scatter.legend_elements(), title = 'Target');

# Add a horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle = '--')
```

```python
heart_disease.head()
```

```python
fig, ax = plt.subplots(figsize = (10, 5))

#plot the data
scatter = ax.scatter(x = heart_disease['age'],
                     y = heart_disease['trestbps'],
                     c = heart_disease['sex'])

#customize

ax.set(title = 'heart disease',
       xlabel = 'age',
       ylabel = 'trestbps')

#add a legend

ax.legend(*scatter.legend_elements(), title = 'sex')

#add a horizontal line

ax.axhline(y = heart_disease['trestbps'].mean(),
           linestyle = '--');
```

```python
#subplot  of chol, age and thalach

fig, (ax1, ax2) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize = (10, 10),
                               sharex = True)

scatter = ax1.scatter(x = over_50['age'],
                      y = over_50['chol'],
                      c = over_50['target'],
                      cmap = 'summer')

ax1.set(title = 'Heart Disease and Cholesterol Levels',
        ylabel = 'Cholesterol')

ax1.legend(*scatter.legend_elements(), title = 'target')

ax1.axhline(y = over_50['chol'].mean(), linestyle = '--')

#axes 2
scatter = ax2.scatter(x = over_50['age'],
                      y = over_50['thalach'],
                      c = over_50['target'],
                      cmap = 'plasma')

ax2.set(title = 'Heart Disease and Max Heart Rates',
        xlabel = 'Age',
        ylabel = 'Thalach')

ax2.legend(*scatter.legend_elements(), title = 'target')
ax2.axhline(y = over_50['thalach'].mean(), linestyle = '--');

#Add a title to the figure
fig.suptitle('Heart Disease Analysis', fontsize = 15, fontweight = 'bold')

```
